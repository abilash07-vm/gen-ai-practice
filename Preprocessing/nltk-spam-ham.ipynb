```python
import pandas as pd
training_data = pd.read_csv('./data-set/spam_ham_dataset.tsv', sep='\t', names=['label', 'message'], skiprows=1)
```

```python
training_data.shape
```

```python
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
import nltk

nltk.download('stopwords')

porter_stemmer = PorterStemmer()

all_stop_words = set(stopwords.words('english'))

```

```python
cleaned_data = []
for sentence in training_data.message:
    words = word_tokenize(re.sub('[^a-zA-Z]',' ', sentence.lower()))
    cleaned_sentence = [porter_stemmer.stem(word) for word in words if word not in all_stop_words]
    cleaned_data.append(' '.join(cleaned_sentence))
print(cleaned_data)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_features=10)
vectors = vectorizer.fit_transform(cleaned_data)
```
