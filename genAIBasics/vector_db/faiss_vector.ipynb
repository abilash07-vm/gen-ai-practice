```python
from langchain_community.vectorstores import FAISS

from langchain_community.document_loaders import TextLoader

txt_loader = TextLoader('../../sample_files/large_files/apj_abdul_kalam_real_1mb.txt')
data = txt_loader.load()
```

```python
from langchain_community.embeddings import OllamaEmbeddings

embedding = OllamaEmbeddings(model='embeddinggemma:latest')
```

```python
from langchain_text_splitters import CharacterTextSplitter

text_splitter = CharacterTextSplitter(chunk_size = 500, chunk_overlap =50)
```

```python
splitted_data = text_splitter.split_documents(data)
```

```python
vector_store = FAISS.from_documents(splitted_data, embedding)
```

```python
query = 'How many unique nodes are there and list it out'

vector_store.similarity_search(query)
```

```python
vector_store.save_local(folder_path= 'faiss_index')
```

```python
saved_db = FAISS.load_local(folder_path= 'faiss_index', embeddings=embedding, allow_dangerous_deserialization=True)
```

```python
query = 'About ABJ abdul kalam\'s Education?'
saved_db._similarity_search_with_relevance_scores(query)
```
