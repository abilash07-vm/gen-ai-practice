```python
import os
from dotenv import load_dotenv

load_dotenv()

os.environ['HF_TOKEN'] = os.getenv('HF_TOKEN')
```

```python
from langchain_huggingface import HuggingFaceEmbeddings

hf_embeding = HuggingFaceEmbeddings(model='all-MiniLM-L6-v2')
```

```python
text = "This is a test for huggingface embeddings"

embeded_text  = hf_embeding.embed_query(text)
```

```python
len(embeded_text)
```

```python
from langchain_community.document_loaders import TextLoader

txt_loader = TextLoader('../../sample_files/text/book_chapter_excerpt.txt')
data = txt_loader.load()
```

```python
embeded_docs = hf_embeding.embed_documents([doc.page_content for doc in data])
```
